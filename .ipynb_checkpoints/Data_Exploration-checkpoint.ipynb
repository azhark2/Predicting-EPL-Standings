# Data Exploration  
**Goal: to visualize our features**  

```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
#define the seasons that we have
seasons = ['2009-2010', '2010-2011', '2011-2012', '2013-2014', '2014-2015', '2015-2016']

#we'll store each season's data as an element in the list datasets
datasets = []

#define file path
filepath = ''

#now loop through the files and add each DataFrame to dataset
for year in seasons:
    filename = filepath + 'team_attributes_processed-' + year + '-raw_data.csv'
    datasets.append(pd.read_csv(filename, index_col = 0))
```

```python
colors = ['red', 'blue', 'green', 'black', 'purple', 'orange']
for j in range(0, len(seasons)):
    
    #get the feature data for a season
    data = datasets[j].iloc[:,2:-1].copy()
    
    #get the score data for a season
    score = datasets[j].iloc[:,-1].copy()

    #initialize figure
    plt.figure(figsize = (20,18))

    #now loop through each feature and plot it against the score
    for i in range(0,24):
        #create subplots - each subplot is the score vs. feature
        plt.subplot(5,5, i+1)
        plt.scatter(data.iloc[:,i], score, c = colors[j])
        plt.title(data.iloc[:,i].name)
        
        #add season info
        if i == 10:
            plt.ylabel(seasons[j] + ' Season\nTeam Scores ', fontsize = 24)
    
#plt.savefig('attributes_plot.jpg')
```
