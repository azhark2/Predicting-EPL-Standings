```python
#import packages
from bs4 import BeautifulSoup
import pandas as pd
import requests
import urllib
import random
import re
import time
```

```python
#this is our url of interest (2016/2017 season)
# url = 'https://www.premierleague.com/tables?co=1&se=54&mw=-1&ha=-1'
urls = ['https://www.premierleague.com/tables?co=1&se=54&mw=-1&ha=-1',
'https://www.premierleague.com/tables?co=1&se=42&mw=-1&ha=-1',
'https://www.premierleague.com/tables?co=1&se=27&mw=-1&ha=-1',
'https://www.premierleague.com/tables?co=1&se=22&mw=-1&ha=-1',
'https://www.premierleague.com/tables?co=1&se=21&mw=-1&ha=-1',
'https://www.premierleague.com/tables?co=1&se=20&mw=-1&ha=-1']
teams = []
season = []
points = []
num_season = 1
for url in urls:
    #get the url information and store it in r
    r = requests.get(url)

    #get the html text from the website
    data = r.text

    #initiate a BeautifulSoup object with our html data
    soup = BeautifulSoup(data, 'html.parser')
    #loop through the first 20 entries (the scraper pulls information from First Team, PL2, U18)
    for i in range(0,20):
        #season number: most recent is 1
        season.append(num_season)
        #this gives us the team name
        teams.append(str(soup.find_all('span', class_ = 'long')[i].text))
        #this gives us the correspoding score
        points.append(int(soup.find_all('td', class_ = 'points')[i].text))
    num_season += 1
```

```python
pd.DataFrame({'Team': teams, 'Score': points, 'Season': season})
```
